## Pretrained Emotion Detector
Facial Expression Recognition Library is developed by Justin Shenk.<br>This Library requires OpenCV>=3.2 and Tensorflow>=1.7.0 dependencies installed in the system.<br> Faces are detected using OpenCV’s Haar Cascade classifier or the more accurate MTCNN.<br> The CNN model used to recognize facial expression is presented in the paper "Real-Time Convolutional Neural Networks for Emotions and Gender Classification" written by O. Arriaga, P. G. Ploger and M. Valdenegro.

### Imports

```python
from fer import Video
from fer import FER
import os
import sys
import pandas as pd
```

### Model preparation and execution

```python
# Location of the video that has to be processed
video_location = "emotion_benchmark.mov"
# Face detector model is loaded (MTCNN is used)
face_detector = FER(mtcnn=True)
# Input the video for processing
input_video = Video(video_location)
```

```python
# The analyze() function will run analysis on every frame of the input video. 
# It will create a bounding box around every face and show the emotion confidence values next to them.
# Finally, the method will publish a video of the result.
processing_data = input_video.analyze(face_detector, display=False)
```

### Results and analysis

```python
# Emotions detected are converted into a dataframe.
vid_df = input_video.to_pandas(processing_data)
vid_df = input_video.get_first_face(vid_df)
vid_df = input_video.get_emotions(vid_df)

# A plot of the emotions detected frame by frame is shown.
pltfig = vid_df.plot(figsize=(20, 8), fontsize=16).get_figure()
```

```python
# Emotions score are summed to show the more frequent emotions in the video.
angry = sum(vid_df.angry)
disgust = sum(vid_df.disgust)
fear = sum(vid_df.fear)
happy = sum(vid_df.happy)
sad = sum(vid_df.sad)
surprise = sum(vid_df.surprise)
neutral = sum(vid_df.neutral)

emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
emotions_values = [angry, disgust, fear, happy, sad, surprise, neutral]

score_comparisons = pd.DataFrame(emotions, columns = ['Human Emotions'])
score_comparisons['Emotion Value from the Video'] = emotions_values
score_comparisons
```
